# Conditional selection (boolean arrays)

We saw conditional selection applied to Series and it'll work in the same way for DataFrames. After all, a DataFrame is a collection of Series:

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Population': [35.467, 63.951, 80.94 , 60.665, 127.061, 64.511, 318.523],
    'GDP': [
        1785387,
        2833687,
        3874437,
        2167744,
        4602367,
        2950039,
        17348075
    ],
    'Surface Area': [
        9984670,
        640679,
        357114,
        301336,
        377930,
        242495,
        9525067
    ],
    'HDI': [
        0.913,
        0.888,
        0.916,
        0.873,
        0.891,
        0.907,
        0.915
    ],
    'Continent': [
        'America',
        'Europe',
        'Europe',
        'Europe',
        'Asia',
        'Europe',
        'America'
    ]
}, columns=['Population', 'GDP', 'Surface Area', 'HDI', 'Continent'])
```

```python
df.index = [
    'Canada',
    'France',
    'Germany',
    'Italy',
    'Japan',
    'United Kingdom',
    'United States',
]
```

```python
df
```

```python
df['Population'] > 70
```

The boolean matching is done at Index level, so you can filter by any row, as long as it contains the right indexes. Column selection still works as expected:

```python
df.loc[df['Population'] > 70, 'Population'] # return only elements which Pops are greater than 70 in the Population colum
```

```python
df.loc[df['Population'] > 70, ['Population', 'GDP']] # returns all elements which Pops are greater than 70  in the Pop and GDP columns
```

# Dropping Stuff

Opposed to the concept of selection, we have "dropping". Instead of pointing out which values you'd like to select you could point which ones you'd like to drop:

```python
df.drop('Canada')
```

```python
df.drop(['Canada', 'Japan'])
```

```python
df.drop(columns=['Population', 'HDI'])
```

```python
df.drop(['Italy', 'Canada'], axis=0) # shows everyting in this row
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis='columns') # returns all row elements shown in columns variable
```

```python
df.drop(['Canada', 'Germany'], axis='rows') # returns all elements shown in rows variable
```

All these drop methods return a new DataFrame. If you'd like to modify it "in place", you can use the inplace attribute (there's an example below).

## Broadcasting Series across a DataFrame

Operations with Series work at a column level, broadcasting down the rows (which can be counter intuitive).

```python
crisis = pd.Series([-1_000_000, -0.3], index=['GDP', 'HDI'])
crisis
```

```python
df[['GDP', 'HDI']]
```

```python
df[['GDP', 'HDI']] + crisis # broadcasts crisis[0] + every element in 1st column of df, broadcasts crisis[1] + every element in 2nd column of df
```

# Modifying DataFrames

It's simple and intuitive, You can add columns, or replace values for columns without issues:

### Remember all operations are normally immutable when working w/ Pandas DataFrames

## Adding a new column

```python
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
langs
```

```python
df['Language'] = langs # adds entire series to dataframe as a new column
```

```python
df
```

## Replacing values per column

```python
df['Language'] = 'English' # changes all values to English
```

```python
df
```

## Renaming Columns

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

```python
df
```

## Dropping columns

```python
df.drop(columns='Language', inplace=True)
```

```python
df
```

Adding values

```python
df.append(pd.Series({
    'Population': 3,
    'GDP': 5
}, name='China'))
```

Append returns a new DataFrame:

We can use drop to just remove a row by index:
